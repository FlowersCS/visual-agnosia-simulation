```python
import matplotlib.pyplot as plt
```

```python
resnet50 = {
    "base": 0.90883,
    "initial": [0.20833, 0.04466, 0.04233, 0.03366],
    "final":   [0.48399, 0.15666, 0.04216, 0.03333]
}

ViT = {
    "base": 0.90850,
    "initial": [0.8858333, 0.8884999, 0.540666, 0.1043333],
    "final":   [0.7636666, 0.6691666, 0.484166, 0.2936666]
}
```

```python
percentages = [0, 20, 30, 50, 70]

resnet50_initial = [resnet50["base"]] + resnet50["initial"]
vit_initial = [ViT["base"]] + ViT["initial"]

plt.figure(figsize=(8, 6))
plt.plot(percentages, vit_initial, label="ViT", color="blue", linewidth=1.5, marker="o")
plt.plot(percentages, resnet50_initial, label="ResNet50", color="red", linewidth=1.5, marker="o")

plt.xlabel("% weight diminution")
plt.ylabel("accuracy")
plt.title("Capas iniciales")
plt.legend()

plt.grid(visible=True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
resnet50_final = [resnet50["base"]] + resnet50["final"]
vit_final = [ViT["base"]] + ViT["final"]

plt.figure(figsize=(8, 6))
plt.plot(percentages, vit_final, label="ViT", color="blue", linewidth=1.5, marker="o")
plt.plot(percentages, resnet50_final, label="ResNet50", color="red", linewidth=1.5, marker="o")

plt.xlabel("% weight diminution")
plt.ylabel("accuracy")
plt.title("Capas finales")
plt.legend()

plt.grid(visible=True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
